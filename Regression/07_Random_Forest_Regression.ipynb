# **Random Forest Regression**




## **Data Preprocessing**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/Position_Salaries.csv')
X = dataset.iloc[: , 1:-1].values
Y = dataset.iloc[: , -1].values
```

## **Training the Dicision Tree model on the whole dataset**

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X, Y)
```

## **Predict on a new result**

it is not the best model to use on single feature dataset

```python
regressor.predict([[6.5]])
```

## **Visualizing the Dicision Tree Regression Results**

it is not good as the Random Forest Regression model is adapted to deal with many features dataset

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X , Y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Dicision Tree Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
