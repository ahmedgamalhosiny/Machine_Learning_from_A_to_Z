**Regression** branch of Machine Learning (ML) that aims to predict some continues real numbers like (salary , temprature , ...)



**Simple linear regression** is the simplest ML model

# **Simple Linear Regression**

## **Data Preprocessing**

```python
import pandas as pd

dataset_path = '/content/Salary_Data.csv'
dataset = pd.read_csv(dataset_path)
print(dataset.head())
```

```python
print(dataset.isna().sum())
```

**reshape(rows, cols)** ▶ changes the shape (dimensions) of an array without changing its data.(total number of elements must remain the same)

(**-1**) tells NumPy: “figure out this dimension automatically”  , Useful when you don’t want to manually calculate rows or columns


```python
import numpy as np
X = dataset['YearsExperience'].to_numpy().reshape(-1, 1)
Y = dataset['Salary'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(X, Y, test_size=0.2, random_state=42)
```

## **Training and Predicting**

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

## **Visualization of Train and Test Results**

**Scatter** ▶ Plot individual points (x, y) on a 2D plane, Each point is independent , No lines are drawn between points.


**Plot** ▶ Connect points with lines, optionally with markers, Points are connected in order by lines.

### **Train**

```python
import matplotlib.pyplot as plt
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, regressor.predict(x_train), color='blue')
plt.title('Salary vs Experiences (Training Set)')
plt.xlabel('Years Of Experience')
plt.ylabel('Salary')
plt.show()
```

### **Test**

**Do we have to reeplace x_train -> x_test in plotting the Regression Line?**

**NO**, remembere that the regression line that we get is  resulted from unique equation , so the predicted salary of the test set is will be on the same regression line **as** the predicted salary of the train set

```python
plt.scatter(x_test, y_test, color='red')
plt.plot(x_train, regressor.predict(x_train), color='blue') # do we have to reeplace x_train -> x_test?  (LOOK AT THE ABOVE COMMENT)
plt.title('Salary vs Experiences (Test Set)')
plt.xlabel('Years Of Experience')
plt.ylabel('Salary')
plt.show()
```
