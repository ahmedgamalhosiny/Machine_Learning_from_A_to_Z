**Polynomial Linear Regression:** is non linear model ,because it can learn non-linear correlations

```
y = b0 + b1x1 + b2(x2)^2 + b3(x3)^3 .......
```

## **Data Preprocessing**

````python
import numpy as np
import pandas as pd

dataset_path ='/content/Position_Salaries.csv'
dataset = pd.read_csv(dataset_path)
print(dataset.head(5))
````

````python
X = dataset.iloc[:, 1:-1].to_numpy()
Y = dataset.iloc[:, -1].to_numpy()
````


## **Training the Linear Regression model on the whold dataset**



````python
from sklearn.linear_model import LinearRegression
lin_regressor = LinearRegression()
lin_regressor.fit(X, Y)
````

## **Training the Polynomial Regression model on the whold dataset**

**first:** create matrix of powered features(x1 , x1^2 , x1^3, ..... , x1^n) using **PolynomialFeatures**

**then** train **Linear Regression model** on the new feature matrix

````python
from sklearn.preprocessing import PolynomialFeatures
poly_regressor = PolynomialFeatures(degree=4) #this degree creates matrix of x1, x1^2
X_poly = poly_regressor.fit_transform(X)

lin_regressor_2 = LinearRegression()
lin_regressor_2.fit(X_poly, Y )
````

## **Visualizing the Linear Regression results**

````python
import matplotlib.pyplot as plt
plt.scatter(X, Y, color='red')
plt.plot(X, lin_regressor.predict(X), color='blue')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
````

## **Visualizing the Polynomial Regression results**

````python
plt.scatter(X, Y, color='red')
plt.plot(X, lin_regressor_2.predict(poly_regressor.fit_transform(X)), color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
````

## **Predicting a new result with Linear Regression**

````python
lin_regressor.predict([[6.5]]) # this give bad result
````

## **Predicting a new result with Polynomial Regression**

````python

lin_regressor_2.predict(poly_regressor.fit_transform([[6.5]])) # this give good result
````
