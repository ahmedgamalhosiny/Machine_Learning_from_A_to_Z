## **Handle Missing Values**



> Implemented well there!
***üòµ check Titanic dataset Preprocessing***


**How you handle missing values depends on:**üó£

* Data type (numerical vs categorical)

* % of missing values

* Whether ‚Äúmissing‚Äù has meaning

* Model type


**__________________________________________________________________**


üòâ**Numerical Columns (Num)**

üîπ 1. Mean / Median Imputation (Most common)

Mean ‚Üí normal distribution

Median ‚Üí skewed data / outliers

```
dataset['Age'] = dataset['Age'].fillna(dataset['Age'].median())
```

üìå Best practice: use median most of the time.

üîπ 2. Drop rows (only if very few missing)

```
dataset.dropna(subset=['Age'], inplace=True)
```


‚úî OK if < 5% missing
‚ùå Bad if many rows lost

üîπ 3. Model-based imputation (Advanced)

```
KNN imputer

Regression imputation

from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
dataset[['Age']] = imputer.fit_transform(dataset[['Age']])
```

üîπ 4. Flag missing values (Very useful ‚≠ê)
```
dataset['Age_missing'] = dataset['Age'].isna().astype(int)
```

Helps the model learn that ‚Äúmissing‚Äù matters.

**____________________________________________________________________**

üòâ**Categorical Columns (Cat)**


üîπ 1. Treat missing as its own category ‚≠ê (Best)
```
dataset['Cabin'] = dataset['Cabin'].fillna('Unknown')
```

‚úî Very common
‚úî Preserves information

üîπ 2. Mode imputation (Only if few missing)
```
dataset['Embarked'] = dataset['Embarked'].fillna(
    dataset['Embarked'].mode()[0]
)
```

‚úî If < 5‚Äì10% missing
‚ùå Not for high missing %

üîπ 3. Drop column (if too many missing)
```
dataset.drop(columns=['ColumnName'], inplace=True)
```

‚úî If > 50% missing
‚úî If no business meaning

üîπ 4. Group rare categories
```
freq = dataset['Deck'].value_counts(normalize=True)
dataset['Deck'] = dataset['Deck'].where(freq > 0.05, 'Other')
```

**____________________________________________________________________**

````python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values= np.nan, strategy='mean' )
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
````

````python
print(X)
````

## **Encoding Categorical Data**

**One Hot Encoding** : turn category column to 'n' columns (n is dependend on the number of different values in column) By creating binary (0,1) vectors for each value (from n values)

why not just make them one column and give each categorical value a specific numerical valueüö© ??

‚ñ∂ as the madel may interpret that as ordering matter and this would mislead the model while trainig

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers= [('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
````

````python
print(X)
````

## **Encoding Independent Column**




**Label Encoding** ‚ñ∂ give each value in independent column a number (class) '0,1,2,3,..' and treat them as numbers then make swap between each number the class represent it

````python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
````

````python
print(Y)
````

**Train Set** part of dataset (80% mostly) used to train the models

**Test Set** the rest of dataset(20%) that is new to the model , we apply the model on it to know to which degree the model generalize and could make right predictions on unseen data

## **Feature Scaling**

**why use scaling??** üîΩ

to make all values in each column in specific to some range of values to avoid that training process affected with column that have high values





we apply **Scaling** in 2 ways:

**Normalization**: used when features are normally distributed

üÜö

**Standarization**: used most of time (we will use it)


**Standarization** :
we fit the Scaler on training set then use it to transform the **training set** **x_train** then transform the **test set** **x_test**

‚ùå do not train the scaler on the test set

````python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# we fit and transform the train values but only apply transformation on the test values with the same scaler "that is fitted on train set"
x_train[:, 3:] = sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] = sc.transform(x_test[:, 3:])
````

````python
print(x_train)
````

````python
print(x_test)
````
